```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma", "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ", "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA", "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ", "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

cleaned_students = []

for student in raw_students:
    clean_name = student["name"].strip().title()
    clean_roll = int(student["roll"])

    marks_parts = student["marks_str"].split(", ")
    clean_marks = []

    for mark in marks_parts:
        clean_marks.append(int(mark))

    words = clean_name.split()
    is_valid = True

    for word in words:
        if not word.isalpha():
            is_valid = False

    print("================================")
    print(f"Student : {clean_name}")
    print(f"Roll No : {clean_roll}")
    print(f"Marks   : {clean_marks}")

    if is_valid:
        print("✓ Valid name")
    else:
        print("✗ Invalid name")

    print("================================")

    cleaned_student = {
        "name": clean_name,
        "roll": clean_roll,
        "marks": clean_marks
    }

    cleaned_students.append(cleaned_student)
```

```python
new_subject_count = 0

while True:
    subject_input = input("Enter subject name: ")

    if subject_input.lower() == "done":
        break

    marks_input = input("Enter marks (0-100): ")

    if not marks_input.isdigit():
        print("Invalid input. Enter numbers only.")
        continue

    mark_value = int(marks_input)

    if mark_value < 0 or mark_value > 100:
        print("Marks must be between 0 and 100.")
        continue

    subjects.append(subject_input)
    marks.append(mark_value)

    new_subject_count += 1
    print("Added successfully!")
```

```python
class_data = [
    ("Ayesha Sharma", [88, 72, 95, 60, 78]),
    ("Rohit Verma", [55, 68, 49, 72, 61]),
    ("Priya Nair", [91, 85, 88, 94, 79]),
    ("Karan Mehta", [40, 55, 38, 62, 50]),
    ("Sneha Pillai", [75, 80, 70, 68, 85]),
]

pass_count = 0
fail_count = 0
topper_name = ""
topper_average = 0
all_averages = []

print("Name              | Average | Status")
print("----------------------------------------")

for student in class_data:
    name, marks_list = student
    average = round(sum(marks_list) / len(marks_list), 2)
    all_averages.append(average)

    if average >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    if average > topper_average:
        topper_average = average
        topper_name = name

    print(f"{name:<17} | {average:<7} | {status}")

class_average = round(sum(all_averages) / len(all_averages), 2)

print()
print("Passed:", pass_count)
print("Failed:", fail_count)
print("Class Topper:", topper_name, topper_average)
print("Class Average:", class_average)
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

clean_essay = essay.strip()
print("1. Stripped essay:")
print(clean_essay)

print("\n2. Title Case:")
print(clean_essay.title())

print("\n3. Count of 'python':")
print(clean_essay.count("python"))

print("\n4. Replace 'python' with 'Python 🐍':")
print(clean_essay.replace("python", "Python 🐍"))

sentences = clean_essay.split(". ")
print("\n5. Sentences list:")
print(sentences)

print("\n6. Numbered sentences:")
for i in range(len(sentences)):
    sentence = sentences[i]

    if not sentence.endswith("."):
        sentence = sentence + "."

    print(f"{i + 1}. {sentence}")
```
